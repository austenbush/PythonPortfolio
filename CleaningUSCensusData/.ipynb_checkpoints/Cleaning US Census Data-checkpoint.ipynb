# Cleaning US Census Data

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import glob
import seaborn as sns
```

```python
files = glob.glob('state*.csv')
df_list = []
for i in files:
    df = pd.read_csv(i)
    df_list.append(df)

df = pd.concat(df_list)
print(df.head())
```

3. Look at the `.columns` and the `.dtypes` of the `us_census` DataFrame. Are those datatypes going to hinder you as you try to make histograms?

```python
df = df.drop(columns=['Unnamed: 0'])
```

```python
df = df.replace('\%', '', regex=True)
```

```python
for i in df.columns[2:8]:
    df[i] = pd.to_numeric(df[i])
```

```python
df['Income'] = df.Income.replace('[\$,\.]', '', regex=True)
```

```python
df['Income'] = pd.to_numeric(df['Income'])
```

```python
df['GenderPop'] = df.GenderPop.replace('M|F', '', regex=True)
gend_pop_split = df.GenderPop.str.split('_')
```

```python
print(gend_pop_split)
df['num_Male'] = gend_pop_split.str.get(0)
df['num_Female'] = gend_pop_split.str.get(1)
```

```python
df = df.drop(columns=['GenderPop'])
```

```python
df['num_Male'] = pd.to_numeric(df.num_Male)
df['num_Female'] = pd.to_numeric(df.num_Female)
```

```python
print(df.head())
```

```python
df = df.drop_duplicates()
```

```python
df.State.unique()
```

```python
df['num_Female'] = df.num_Female.fillna(0)
```

```python
num_fem_funct = lambda row: row['TotalPop']-row['num_Male'] if row['num_Female'] == 0 else row['num_Female']
```

```python
df['num_Female'] = df.apply(num_fem_funct, axis=1)
```

```python
print(df.columns[2:7])
```

```python
df['Pacific'] = df.Pacific.fillna(0)
df['Pacific'] = df.apply(lambda row: 100-np.sum([row[i] for i in df.columns[2:7]]), axis=1)
```

```python
df.info()
```

```python
df.describe()
```

```python
sns.lmplot(data=df, x='num_Female', y='Income')
plt.show()
plt.clf()
```

```python
for i in df.columns[2:8]:
    sns.lmplot(x=i, y='Income', data=df)
    plt.show()
    plt.clf()
```

```python
sns.scatterplot(x='TotalPop', y='Income', data=df)
plt.show()
plt.clf()
```
