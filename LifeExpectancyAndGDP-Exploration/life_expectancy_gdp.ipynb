## Life Expectancy Vs GDP
An exploration in to the relationship between life expectancy and GDP.

```python
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv('all_data.csv')
df = df.rename(columns={'Life expectancy at birth (years)': 'Life_Expectancy'})
```

```python
print(df.head())
```

```python
plt.figure(figsize=(8,6))
sns.scatterplot(x='GDP', y='Life_Expectancy', data=df, hue='Country')
plt.savefig('LE_GDP_all.png')
plt.show()
plt.clf()
```

At higher GDPs, there is a positive relationship between life expectancy and GPD for a country. However, the relationship is much stronger at low GDPs.<br><br>
In other words, GDP growth has a larger effect on life expectancy in countries that have a lower overall GDP.

```python
for i in df.Country.unique():
    n_df = df[df.Country == i]
    sns.lmplot(x='GDP', y='Life_Expectancy', data=n_df)
    plt.title('GDP vs Life Expectancy in ' + i)
    plt.xlabel('GDP')
    plt.ylabel('Life Expectancy')
    plt.show()
    plt.close()
```

The above graphs further illustrate the positive impact between life expectancy and GDP. Every country represented in the data shows a positive correlation.

```python
for i in df.Country.unique():
    n_df = df[df.Country == i]
    plt.subplot(1,2,1)
    plt.plot(n_df.Year, n_df.Life_Expectancy)
    plt.title('Life Expectancy by Year in ' + i)
    plt.xlabel('Year')
    plt.ylabel('Life Expectancy')
    plt.subplot(1,2,2)
    plt.plot(n_df.Year, n_df.GDP)
    plt.title('GDP by Year in ' + i)
    plt.xlabel('Year')
    plt.ylabel('GDP')
    plt.subplots_adjust(wspace = 0.35, hspace = 0.5)
    plt.show()
    plt.clf()
```

#### Conclusions
1. GDP and Life expectancy are positively correlated
2. Lower life expectancy/GDP combos show a faster rate of increase against a similar percentage increase in GDP.
3. Highest GDP does not necessarily equal highest life expectancy. However, increasing GDP does equal increasing life expectancy.
