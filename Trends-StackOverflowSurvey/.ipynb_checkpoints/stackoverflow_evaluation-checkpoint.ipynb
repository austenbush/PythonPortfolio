Below we calculate the p-val of variation in the number of job seekers due to the type of degree. In this test, we use a confidence level of 0.05 (95% chance of the variation being statistically significant).<br><br>
In this case, we are using a two-tailed test. If we were to use a one-tailed test, statistic significance would increase.

```python
population_mean = len(df[df.jobseek == 'I am actively looking for a job'])/len(df)
for i in df.undergradmajor.unique():
    pop_len = len(df[df.undergradmajor == i])
    samp = df
    samp_count = len(df[(df.jobseek == 'I am actively looking for a job') & (df.undergradmajor == i)])
    prob_not_mean = stats.binomtest(samp_count, n=pop_len, p=population_mean)
    if prob_not_mean.pvalue < 0.05:
        print(i + ' degree for job seekers is statistically significant with a value of ' + str(round(prob_not_mean.pvalue,4)))
```

```python
population_mean = len(df[df.employment == 'Not employed, but looking for work'])/len(df)
for i in df.undergradmajor.unique():
    pop_len = len(df[df.undergradmajor == i])
    samp = df
    samp_count = len(df[(df.employment == 'Not employed, but looking for work') & (df.undergradmajor == i)])
    prob_not_mean = stats.binomtest(samp_count, n=pop_len, p=population_mean)
    if prob_not_mean.pvalue < 0.05:
        print(i + ' degree for those unemployed but looking is statistically significant with a value of ' + str(round(prob_not_mean.pvalue,4)))

print(population_mean)
```

```python
pop_len = len(df[df.undergradmajor == 'Computer science, computer engineering, or software engineering'])
samp = df
samp_count = len(df[(df.jobseek == 'I am actively looking for a job') & (df.undergradmajor == 'Computer science, computer engineering, or software engineering')])

print(samp_count/pop_len)
```

```python
year_comp_df = df[['yearscodepro', 'convertedcomp']]
#traindf, testdf = train_test_split(year_comp_df, train_size = 0.2)
imp = IterativeImputer(max_iter=20, random_state=0)
imp.fit(year_comp_df)
compdf = pd.DataFrame(np.round(imp.transform(year_comp_df),0), columns=['yearscodepro', 'convertedcomp'])
```

```python
compPlotdf = compdf.loc[compdf['convertedcomp'] <= 150000]
compPlotdf['CodeYearBins'] = pd.qcut(compPlotdf['yearscodepro'], q=5)

sns.boxplot(x="CodeYearBins", y="convertedcomp",
            data=compPlotdf)
```
